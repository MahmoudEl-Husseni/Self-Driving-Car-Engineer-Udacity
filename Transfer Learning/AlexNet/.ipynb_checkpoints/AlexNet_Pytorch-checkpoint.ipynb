```python
import pickle
import warnings
import cv2 as cv
import numpy as np
import pandas as pd


import torch 
import torchvision
from torch import nn
from torchsummary import summary
from torchvision import transforms
from torch.utils.data import DataLoader

from sklearn.preprocessing import OneHotEncoder

warnings.simplefilter('ignore')
config = {
    'LR'                : 1e-3, 
    'EPOCHS'            : 10,
    'TEST_BATCH_SIZE'   : 128,
    'TRAIN_BATCH_SIZE'  : 256,
}
```

```python
label_names = pd.read_csv('CarND-Alexnet-Feature-Extraction/signnames.csv', index_col='ClassId')

with open('train.p', 'rb') as f:
  train = pickle.load(f)
from torchvision import models as models

model = models.alexnet(pretrained=True)

# freeze all feature extraction (convolution) layers
for param in list(model.parameters())[:-6]:
  param.requires_grad = False

model.classifier[6] = nn.Linear(in_features=(4096), out_features=43, bias=True)

display(label_names.head())
display(summary(model, input_size=(3, 64, 64)))
```

```python
preprocessed_train = [cv.resize(im, (64, 64)) for im in train['features']]
preprocessed_train -= np.mean(preprocessed_train, 0)
preprocessed_train /= np.std(preprocessed_train, 0)
```

```python
import matplotlib.pyplot as plt
def visualize_images(rows, cols, images, titles):
  fig, axs = plt.subplots(rows, cols, figsize=(20, 12))
  for i, ax in enumerate(axs.flatten()):
    ax.imshow(images[i], cmap='gray')
    ax.set_title(titles[i])
    ax.set_xticks([]), ax.set_yticks([]);

rows, cols = 4, 4 
indices = np.random.randint(0, len(train['features']), rows*cols)
imagestoplot = preprocessed_train[indices].reshape(-1, 64, 64, 3)
classes = train['labels'][indices]
titles = [label_names.iloc[_class, 0] for _class in classes]
visualize_images(rows, cols, imagestoplot, titles);
```

```python
ohe = OneHotEncoder()
ohe.fit(train['labels'].reshape(-1, 1))
coded_labels = ohe.transform(train['labels'].reshape(-1, 1)).toarray()

loss = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), config['LR'])

l = 0.0

for epoch in range(config['EPOCHS']):
  indices = list(range(len(train['features'])))
  np.random.shuffle(indices)
  data = preprocessed_train[indices]
  target = coded_labels[indices]
  
  STEPS = len(train['features']) // config['TRAIN_BATCH_SIZE']
  for step in range(STEPS): 
    LOG = f"{f'[info]: epoch:{epoch+1}'.ljust(20)}|{'='.ljust(step, '=')}>{' '.ljust((STEPS-step))}|\t loss: {l}"
    print("\r", end=LOG)
    optimizer.zero_grad()
    inp = torch.Tensor(data[step*config['TRAIN_BATCH_SIZE']: (step+1)*config['TRAIN_BATCH_SIZE']]).view(-1, 3, 64, 64)
    y_true = torch.Tensor(target[step*config['TRAIN_BATCH_SIZE']: (step+1)*config['TRAIN_BATCH_SIZE']])
    y_pred = model.forward(inp)

    l = loss(y_pred, y_true)
    optimizer.step()
  print('')
```
